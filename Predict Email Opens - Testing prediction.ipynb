```python
# Filling 0 for nan

data['last_online'] = data['last_online'].fillna(1.463267e+09)
data['mail_category'] = data['mail_category'].fillna('ffill')
data['mail_type'] = data['mail_type'].fillna('bfill')
data['hacker_timezone'] = data['hacker_timezone'].fillna(5.593259e+03)
data.isna().sum()
```

```python
data
```

```python
# Dummy variables - Label Encoder

from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()

data['mail_category'] = enc.fit_transform(data['mail_category'])
data['mail_type'] = enc.fit_transform(data['mail_type'])
data['hacker_confirmation'] = enc.fit_transform(data['hacker_confirmation'])
```

```python
data.head()
```

```python
data.describe()
```

```python
# Standardization - Scaler with min = 0, max = 1

def norm_fun(i):
    x = (i - i.min())/(i.max() - i.min())
    
    return x
```

```python
data_norm = norm_fun(data)
data = pd.DataFrame(data_norm)
data.describe()
```

```python
data.isna().sum()
```

```python
data = data.fillna(0)
```

```python
data.describe()
```

```python
# Spliting the data

X = data.iloc[:,1:].values
Y = data.iloc[:,0].values
```

```python
# Continous to Discreatization

X = (X > 0).astype(int)
Y = (Y > 0).astype(int)
```

```python
# Model Building

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X, Y)

# Checking the model score

model.score(X, Y)
```

```python
y_pred = model.predict(X)
```

```python
# Confusion Matrix

from sklearn.metrics import confusion_matrix, classification_report
```

```python
cm = confusion_matrix(Y, y_pred)
print(cm)
```

```python
print(classification_report(Y, y_pred))
```
